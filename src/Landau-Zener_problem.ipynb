# Generic eq.

$$
\mathrm{i} \epsilon \frac{\mathrm{d}}{\mathrm{d} \tau}
\begin{pmatrix}
  a \\
  b
\end{pmatrix}
=
\begin{pmatrix}
  \tau & 1 \\
  1  & - \tau
\end{pmatrix}
\begin{pmatrix}
  a \\
  b
\end{pmatrix}
$$
